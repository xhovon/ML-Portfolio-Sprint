```python
import pandas as pd
```

```python
# Create some sample data as a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily','Mop','Mop'],
    'Age': [25, 30, 35, 40, 28,22,22],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'London','Sajek','Sajek'],
    'Salary': [70000, 85000, 90000, 110000, 75000,60000,60000]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)
```

```python
# Combine conditions (use & for 'and')
# Let's find people in London AND older than 29
df[(df['City'] == 'London') & (df['Age'] > 20)]
```

```python
# Let's add a new row with a missing age
new_row = {'Name': 'Frank', 'City': 'Chicago', 'Salary': 50000}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# Look at the data now (you'll see 'NaN' for Frank's age)
df.tail()
```

```python
new_row = {'Name': 'Lori', 'City': 'Chicago', 'Age': 50}
df = pd.concat([df,pd.DataFrame([new_row])], ignore_index=True)
df.tail()
```

```python
df_cleaned = df.dropna()
print(df_cleaned)
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
print("new cleaned dataframe:",df_cleaned)
```

```python
average_age = df['Age'].mean()
avarage_salary = df['Salary'].mean()
df_filled = df.fillna({'Age': average_age,'Salary': avarage_salary}) 
print(df_filled)
```
