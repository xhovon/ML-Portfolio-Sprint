```python
# Copy this into a cell
import io

csv_data = """Name,Age,City,Salary
Alice,25,New York,70000
Bob,30,London,85000
Charlie,35,New York,90000
David,40,,110000
Emily,28,Paris,
Fiona,,London,95000
"""

# This magic line reads a string as if it's a file
df = pd.read_csv(io.StringIO(csv_data))

# Run this to see your starting data
print("--- Initial Data ---")
df.info()
```

```python
df.head()
```

```python
avarage_salary = df['Salary'].mean()
print(f"Average Salary: {avarage_salary}")
```

```python
df[['Name','City']]
```

```python
df[df['City']=='New York']
```

```python
df[df['Age']>30]
```

```python
median_age = df['Age'].median()
print(f"Median Age: {median_age}")  
```

```python
df_filled = df.fillna({'Age': median_age, 'Salary': avarage_salary})
print("--- Data after filling missing values ---")
df_filled.info()
```

```python
df_dropped = df.dropna()
print("--- Data after dropping missing values ---")
df_dropped.info()
```
